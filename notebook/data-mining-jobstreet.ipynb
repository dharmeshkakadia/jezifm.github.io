# Data Scientist Opportunities

Data Mining from various website for data science job postings

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime as dt
```

```python
# Configuration
JOBSTREET_URL = 'http://www.jobstreet.com.ph/en/job-search/search/data-scientist-jobs/'
```

```python
def parse_job_date_posted(job_date_posted):
    job_date_posted = job_date_posted[6:]
    if job_date_posted == 'Yesterday':
        return dt.datetime.today() - dt.timedelta(days=1)
    date_created = (dt.datetime.strptime(job_date_posted, '%d %b - %H:%M %p')
                    .replace(year=2016))
    return date_created
```

```python
with requests.Session() as s:
    soup = BeautifulSoup(s.get(JOBSTREET_URL).text, 'html.parser')
    jobstreet_data = []
    for panel_body in soup.select('.panel-body'):
        position_title = panel_body.select('.position-title')#
        if len(position_title):
            jobstreet_data.append({
                    'position_title': panel_body.select('.position-title')[0].text.strip(),
                    'company_name': panel_body.select('.company-name')[0].text,
                    'job_location': panel_body.select('.job-location')[0].text,
                    'position_title_url': panel_body.select('.position-title-link')[0]['href'],
                    'job_date_posted': panel_body.select('.job-date-posted')[0].text
                })
    jobstreet = pd.DataFrame(jobstreet_data)
    jobstreet['date_created'] = jobstreet['job_date_posted'].apply(parse_job_date_posted)
    jobstreet.sort_values(['date_created'], inplace=True)
```
